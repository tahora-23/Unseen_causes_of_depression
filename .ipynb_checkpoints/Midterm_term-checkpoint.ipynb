```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('democracy-countries-2023.csv')

df2 = pd.read_excel('voilance_rate.xlsx')

df3 = pd.read_csv('literacy-rate-by-country-2023.csv')

df4= pd.read_csv('gender-equality-by-country-2023.csv')

df5 = pd.read_excel('public_service_index.xlsx')

df6= pd.read_excel('houshold_size.xlsx')

df7 = pd.read_csv('unemployment-by-country-2023.csv')

df8 = pd.read_csv('wealth-inequality-by-country-2023.csv')

df9= pd.read_excel('Average_temprature.xlsx')

df10 = pd.read_excel('corruption_perceptions _index_2022.xlsx')

df11 = pd.read_csv('depression-rates-by-country-2023.csv')
```

```python
print(df1.columns)
print(df2.columns)
print(df3.columns)
print(df4.columns)
print(df5.columns)
print(df6.columns)
print(df7.columns)
print(df8.columns)
print(df9.columns)
print(df10.columns)
print(df11.columns)
```

```python
# List of table names
dfs = [df11, df10, df9, df8, df7, df6, df5, df4, df3, df2, df1]
concatenated_df = pd.concat(dfs).groupby('country').first().reset_index()

concatenated_df.columns
```

```python
concatenated_df.columns
```

```python
selected_columns = ['country', 'prevalence', 'corruption_score', 'Average temprature', 'gini', 'rateWb', 'Avg size', 'Public services index', 'score', 'crime index', 'latestRate', 'gendEqal2021', 'education', 'health', ]
```

```python
selected_columns
```

```python
# Use loc to select the columns by name and assign the result back to the DataFrame
df = concatenated_df.loc[:, selected_columns]
df
```

```python
df = df.rename(columns={'score': 'democracy_score', 'latestRate': 'litracy_rate', 'prevalence': 'depression_rate', 'rateWb': 'unemployment_rate', 'Avg size': 'family_avg_size', 'gendEqal2021': 'gender_equality', 'Average temprature': 'average_temperature'})
```

```python
df
```

```python
null_values = df.isnull().sum()
```

```python
null_values
```

```python
df = df.dropna(axis=0)
```

```python
df
```

```python
print(df['average_temperature'])
```

```python
df.dtypes
```

```python
# Remove the '°F' suffix and convert to float
df['average_temperature'] = df['average_temperature'].str.replace(' °F', '').astype(float)
```

```python
df
```

```python
# Convert Fahrenheit to Celsius
df['average_temperature'] = (df['average_temperature'] - 32) * 5/9
```

```python
#Round the values to two decimal places
df['average_temperature'] = df['average_temperature'].round(2)
```
