```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df1 = pd.read_csv('democracy_rate.csv')

df2 = pd.read_csv('crime_rate.csv')

df3 = pd.read_csv('literacy_rate.csv')

df4= pd.read_csv('gender_gap.csv')

df5 = pd.read_csv('public_service_rate.csv')

df6= pd.read_csv('family_size.csv')

df7 = pd.read_csv('unemployment_rate.csv')

df8 = pd.read_csv('wealth_rate.csv')

df9= pd.read_excel('Average_temprature.xlsx')

df10 = pd.read_csv('corruption_rate.csv')

df11 = pd.read_csv('water_scarcity.csv')

df12 = pd.read_csv('air_quality_rate.csv')

df13 = pd.read_csv('depression_rate.csv')
```

```python
# List of table names
dfs = [df13,df10, df9, df8, df7, df6, df5, df4, df3, df2, df1]
concatenated_df = pd.concat(dfs).groupby('country').first().reset_index()

concatenated_df
```

```python
df4
```

```python
merged_df = concatenated_df.merge(df12, on='country')
merged_df
```

```python
selected_columns = ['country','prevalence','corruption_score', 'Average temprature', 'gini', 'rateWb', 'Family size', 'Public services index', 'crime index','latestRate', 'gendEqal2021','score','Air Quality']
```

```python
selected_columns
```

```python
# Use loc to select the columns by name and assign the result back to the DataFrame
df = merged_df.loc[:, selected_columns]
df
```

```python
df = df.rename(columns={'score': 'democracy_score', 'latestRate': 'litracy_rate', 'prevalence': 'depression_rate', 'rateWb': 'unemployment_rate', 'gendEqal2021': 'gender_equality', 'Average temprature': 'average_temperature'})
```

```python
df
```

```python
df['litracy_rate'] = df['litracy_rate'].round(2)
```

```python
df
```

```python
null_values = df.isnull().sum()
```

```python
null_values
```

```python
print(df['average_temperature'])
```

```python
df.dtypes
```

```python
# Remove the '°F' suffix and convert to float
df['average_temperature'] = df['average_temperature'].str.replace(' °F', '').astype(float)
```

```python
df
```

```python
# Convert Fahrenheit to Celsius
df['average_temperature'] = (df['average_temperature'] - 32) * 5/9
```

```python
#Round the values to two decimal places
df['average_temperature'] = df['average_temperature'].round(2)
```

```python
df.reset_index(drop=True, inplace= True)
df
```

```python
df.to_csv('clean_df3.csv', index=True)
unique_values = df['country'].unique()
print(unique_values)
```
